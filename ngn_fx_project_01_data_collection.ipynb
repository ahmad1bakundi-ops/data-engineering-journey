```python
import yfinance as yf
import pandas as pd
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
```

```python
PROJECT_PATH = '/content/drive/MyDrive/data-engineering-journey/ngn-fx-project/'

# Create the folder if it doesn't exist
import os
os.makedirs(PROJECT_PATH, exist_ok=True)

PROJECT_PATH
```

```python
tickers = ['USDNGN=X', 'EURNGN=X', 'GBPNGN=X', 'JPYNGN=X']

data = yf.download(
    tickers,
    start='2015-01-01',
    end='2025-01-01',
    interval='1d',
    progress=False,
    auto_adjust=True
)

data.shape
```

```python
data.head()
```

```python
close = data['Close'].copy()
close.head()
```

```python
close.columns
```

```python
close.shape
```

```python
close.info()
```

```python
close.describe()
```

```python
close.isnull().sum()
```

```python
close[close['USD'] < 10]
```

```python
close[close['GBP'] < 180]
```

```python
close.loc[close['USD'] < 10, 'USD'] = np.nan
close.loc[close['GBP'] < 180, 'GBP'] = np.nan
close.isnull().sum()
```

```python
close = close.interpolate(method='linear')
close.isnull().sum()
```

```python
close[close['USD'] < 150]
```

```python
close[close['GBP'] < 250]
```

```python
# GBP should always be greater than USD historically
# Any row where GBP <= USD is a data error
bad_gbp_mask = close['GBP'] <= close['USD']
close.loc[bad_gbp_mask, 'GBP'] = np.nan
close.isnull().sum()
```

```python
close = close.interpolate(method='linear')
close.isnull().sum()
```

```python
close.describe()
```

```python
close[close['GBP'] <= close['USD']]
```

```python
close.to_csv(PROJECT_PATH + 'ngn_fx_daily_clean.csv')
```
