```python
import pandas as pd
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
```

```python
DATA_PATH = '/content/drive/MyDrive/data-engineering-journey/lesson-02-data/'
df = pd.read_csv(DATA_PATH + 'nigerian_ecommerce_sales_CLEAN.csv')
df.shape
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df.info()
```

```python
df['revenue'] = df['quantity'] * df['unit_price_ngn']
df[['order_id', 'product', 'quantity', 'unit_price_ngn', 'revenue']].head()
```

```python
df.groupby('customer_state')['revenue'].sum()
```

```python
df.groupby('customer_state')['revenue'].sum().sort_values(ascending=False)
```

```python
df.groupby('category')['order_id'].count().sort_values(ascending=False)
```

```python
df.groupby('customer_state')['order_id'].count().sort_values(ascending=False)
```

```python
df.groupby('customer_state')['revenue'].mean().sort_values(ascending=False)
```

```python
df.groupby('customer_state').agg(
    order_count=('order_id', 'count'),
    total_revenue=('revenue', 'sum'),
    avg_order_value=('revenue', 'mean'),
    max_order=('revenue', 'max'),
).sort_values('total_revenue', ascending=False)
```

```python
df.groupby('category').agg(
    order_count=('order_id', 'count'),
    total_revenue=('revenue', 'sum'),
    avg_order_value=('revenue', 'mean'),
    max_order=('revenue', 'max'),
).sort_values('total_revenue', ascending=False)
```

```python
df['order_month'] = df['order_date'].dt.to_period('M')
df.groupby('order_month')['revenue'].sum()
```
