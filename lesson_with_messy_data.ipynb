```python
df_clean['original_currency'] = np.where(
    df_clean['amount'].str.contains('$', regex=False, na=False),  # condition: contains $
    'USD',                                                          # if True: USD
    'NGN'                                                           # if False: NGN
)
df_clean['original_currency']
```

```python
df_clean['amount_naira'] = np.where(
    df_clean['original_currency'] == 'USD',  # condition: is it dollars?
    df_clean['amount_value'] * 1400,          # if True: multiply by exchange rate
    df_clean['amount_value']                  # if False: keep as-is (already naira)
)
df_clean['amount_naira']
```
