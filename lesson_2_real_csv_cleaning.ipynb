```python
df_clean = df.copy()
```

```python
df_clean['customer_state'] = df_clean['customer_state'].str.strip().str.title()
df_clean['customer_state'].value_counts()
```

```python
state_corrections = {
    'Lagos State': 'Lagos',
    'Portharcourt': 'Port Harcourt',
    'Fct': 'Abuja',
    'Benincity': 'Benin City',
    'Benin': 'Benin City',
    'Enugu State': 'Enugu',
    'Ph': 'Port Harcourt',
}

df_clean['customer_state'] = df_clean['customer_state'].replace(state_corrections)
df_clean['customer_state'].value_counts()
```

```python
df_clean['payment_method'].value_counts()
```

```python
df_clean['payment_method'] = df_clean['payment_method'].str.strip().str.title()
df_clean['payment_method'].value_counts()
```

```python
payment_corrections = {
    'Transfer': 'Bank Transfer',
    'Pod': 'Cash On Delivery',
    'Pay On Delivery': 'Cash On Delivery',
    'Cash': 'Cash On Delivery',  # assuming "Cash" means Cash on Delivery in our e-commerce context
}

df_clean['payment_method'] = df_clean['payment_method'].replace(payment_corrections)
df_clean['payment_method'].value_counts()
```

```python
df_clean['status'] = df_clean['status'].str.strip().str.title()
df_clean['status'].value_counts()
```

```python
df_clean['quantity'].value_counts()
```

```python
df_clean['quantity'] = pd.to_numeric(df_clean['quantity'], errors='coerce')
df_clean['quantity']
```

```python
df_clean['quantity'].dtype
```

```python
df_clean['quantity'].isnull().sum()
```

```python
df_clean['order_date'] = pd.to_datetime(df_clean['order_date'], format='mixed', dayfirst=True)
df_clean['order_date']
```

```python
df_clean[['order_id', 'order_date']].head(10)
```

```python
df_clean.info()
```

```python
# Mark impossible prices as NaN
df_clean.loc[
    (df_clean['unit_price_ngn'] <= 0) | (df_clean['unit_price_ngn'] > 500000),
    'unit_price_ngn'
] = np.nan

# Mark impossible quantities as NaN
df_clean.loc[
    (df_clean['quantity'] < 1) | (df_clean['quantity'] > 50),
    'quantity'
] = np.nan
```

```python
df_clean[['unit_price_ngn', 'quantity']].describe()
```

```python
df_clean[df_clean['unit_price_ngn'] > 200000]
```

```python
df_clean.isnull().sum()
```

```python
df_clean = df_clean.dropna(subset=['quantity', 'unit_price_ngn'])
df_clean.shape
```

```python
df_clean.duplicated().sum()
```

```python
df_clean[df_clean.duplicated(keep=False)]
```

```python
df_clean = df_clean.drop_duplicates()
df_clean.shape
```

```python
df_clean.to_csv(DATA_PATH + 'nigerian_ecommerce_sales_CLEAN.csv', index=False)
```
