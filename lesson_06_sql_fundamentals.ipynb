```python
import duckdb
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')
```

```python
DATA_PATH = '/content/drive/MyDrive/data-engineering-journey/lesson-02-data/'

# Create a DuckDB connection
con = duckdb.connect()

# Run our first SQL query!
result = con.execute(f"""
    SELECT *
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv'
    LIMIT 5
""").fetchdf()

result
```

```python
con.execute(f"""
    SELECT order_id, product, unit_price_ngn
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv'
    LIMIT 10
""").fetchdf()
```

```python
con.execute(f"""
    SELECT *
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv'
    WHERE status = 'Delivered'
    LIMIT 10
""").fetchdf()
```

```python
con.execute(f"""
    SELECT order_id, product, unit_price_ngn, status
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv'
    WHERE status = 'Delivered'
      AND unit_price_ngn > 50000
    LIMIT 10
""").fetchdf()
```

```python
con.execute(f"""
    SELECT
        customer_state,
        SUM(quantity * unit_price_ngn) AS total_revenue
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv'
    GROUP BY customer_state
    ORDER BY total_revenue DESC
""").fetchdf()
```

```python
con.execute(f"""
    SELECT
        category,
        AVG(quantity * unit_price_ngn) AS avg_revenue
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv'
    GROUP BY category
    ORDER BY avg_revenue DESC
""").fetchdf()
```

```python
con.execute(f"""
    SELECT *
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv' AS s
    LEFT JOIN '{DATA_PATH}nigerian_ecommerce_customers.csv' AS c
        ON s.customer_id = c.customer_id
    LIMIT 5
""").fetchdf()
```

```python
con.execute(f"""
    SELECT
        c.tier,
        COUNT(*) AS order_count,
        SUM(s.quantity * s.unit_price_ngn) AS total_revenue,
        AVG(s.quantity * s.unit_price_ngn) AS avg_order_value
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv' AS s
    LEFT JOIN '{DATA_PATH}nigerian_ecommerce_customers.csv' AS c
        ON s.customer_id = c.customer_id
    GROUP BY c.tier
    ORDER BY avg_order_value DESC
""").fetchdf()
```

```python
con.execute(f"""
    WITH delivered AS (
        SELECT *
        FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv'
        WHERE status = 'Delivered'
    )
    SELECT
        c.tier,
        AVG(d.quantity * d.unit_price_ngn) AS avg_revenue
    FROM delivered AS d
    LEFT JOIN '{DATA_PATH}nigerian_ecommerce_customers.csv' AS c
        ON d.customer_id = c.customer_id
    GROUP BY c.tier
    ORDER BY avg_revenue DESC
""").fetchdf()
```

```python
con.execute(f"""
    SELECT
        order_id,
        customer_id,
        order_date,
        quantity * unit_price_ngn AS revenue,
        SUM(quantity * unit_price_ngn) OVER (
            PARTITION BY customer_id
            ORDER BY order_date
        ) AS running_total
    FROM '{DATA_PATH}nigerian_ecommerce_sales_CLEAN.csv'
    ORDER BY customer_id, order_date
    LIMIT 20
""").fetchdf()
```
